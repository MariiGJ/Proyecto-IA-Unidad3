```python
images_path = "C:\ProyectoIA_UNIDAD3\ProeyectoIAU3"
class_names = os.listdir(images_path)
class_names.sort()
print(class_names)
```

```python
'''Se realiza un copiado de las imagenes de dataset pricipal y se las manda al nuevo directorio'''
#bucle que se ejecutara hasta que se hayan leido todas las clase
for class_name in class_names:
    # ruta especifica con el nombre de la clase
    cur_path = images_path + "/" + class_name
    # se definen los elementos de la ruta indicada
    elements = os.listdir(cur_path)
    #bucle que se encuarga de leer todos elementos
    for element in elements:
        # Imprimimos la acción realizada
        print("Copying Image - {} in Fotos_Augmented{}".format(element, class_name))
        # se define la ruta de origen de los elementos
        src = os.path.join(images_path + "/" + class_name, element) # origen
         # se define la ruta de destino de los elementos
        dst = os.path.join("C:\ProyectoIA_UNIDAD3\Fotos_Augmented/" + class_name, element) 
        #se copian las imagenes
        shutil.copy(src, dst)
```

```python
datagen = ImageDataGenerator(
        rotation_range=40,
        width_shift_range=0.2,
        height_shift_range=0.2,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.4,
        horizontal_flip=True,
        fill_mode='nearest')
```

```python
'''
    Este método recorre cada uno de los elementos que se conforman en la ruta específicada
    tomando en cuenta la categoría de estos.
    
    Parámetros
    ----------
        categories:
            Indica la categoria a la que pertenece cada una de las rutas
        
    Returns:
        img_lst: 
            Retorna la lista que contiene las imagenes
        labels: 
            Retorna la lista de las etiquetas.
'''
def load_images_and_labels(categories):
    # Instanciamos la lista de imagenes
    img_lst=[]
    # Instanciamos la lista de etiquetas
    labels=[]
    # Bucle que recorre las categorias
    for index, category in enumerate(categories):
        # Bucle que recorre la ruta de las imagenes
        for image_name in os.listdir(fpath+"/"+category):
            # Leer las imagenes
            img = cv2.imread(fpath+"/"+category+"/"+image_name)
            # Transformamos el espacio de color de las imagenes
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            #se establece el tamaño que va a tener cada foto
            IMAGE_SIZE=(192,192)
            #Se realiza el ajuste del tamaño de las imagenes
            img=cv2.resize(img,IMAGE_SIZE)
            # Genera una arreglo con las imagenes y su formato
            img_array = Image.fromarray(img, 'RGB')
            # Añadimos la imagen a la lista
            img_lst.append(np.array(img_array))
            # Agregamos el indice a las etiquetas
            labels.append(index)
    #Retornamos la lista de imagenes y etiquetas
    return img_lst, labels

# Cargamos las imagenes y las etiquetas
images, labels = load_images_and_labels(categories)
# Mostramos la cantidad de imagenes y etiquetas cargadas
print("No. of images loaded = ",len(images),"\nNo. of labels loaded = ",len(labels))
# Mostramos el tipo de arreglo creado
print(type(images),type(labels))
```
